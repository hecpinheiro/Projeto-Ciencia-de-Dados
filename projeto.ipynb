```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
database.isnull().sum()
```

# **Pré-processamento inicial dos dados**

```python
#Remoção da coluna 'id', não é necessária para análise
database.drop('id', axis=1, inplace=True)
database.head(1)
```

```python
#Calculando o média dos valorede de 'bmi', para preencher os campos vazios que a coluna possui
average = database['bmi'].mean()
database['bmi']= database['bmi'].fillna(average)
print("Média de bmi: ",average)
database.isnull().sum()
```

# **Visualização dos dados**

```python
#funções para os plots dos dados numéricos ou categóricos com relação ao 'stroke'
def numericalPlot(var):
  sn.kdeplot(x = database.loc[database['stroke']==1][var], color = 'RoyalBlue', label = 'Stroke', shade = True)
  sn.kdeplot(x = database.loc[database['stroke']==0][var], color = 'MediumAquamarine', label = 'No Stroke', shade = True)
  plt.legend(loc = 'upper right')
  plt.grid(linestyle = '--', axis = 'y')

def categoricalPlot(var):
  sn.countplot(database[var], palette = 'YlGnBu', hue = database['stroke'])
  plt.grid(linestyle =':', axis = 'y')
```

# **Pré-processamento final dos dados**

```python
#o 'Residence_type' não influi nos casos de 'stroke'
database.drop('Residence_type', axis=1, inplace=True)
database.head(1)
```

```python
#convertendo os dados categóricos
database['work_type'] = database['work_type'].map({'Private':0, 'Self-employed': 1, 'Govt_job':2, 'children':3, 'Never_worked':4})
database['gender'] = database['gender'].map({'Male':0, 'Female':1})
database['smoking_status'] = database['smoking_status'].map({'formerly smoked':0, 'never smoked':1, 'smokes':2, 'Unknown':3})
database['ever_married'] = database['ever_married'].map({'No':0, 'Yes':1})

database.head()
```

# **Aprendizado de Máquina**

```python
#separando os dados em 'features' e 'target' para usar nos algoritmos de classificação
features = ['age',
 'hypertension',
 'heart_disease',
 'ever_married',
 'avg_glucose_level',
 'bmi',
 'gender',
 'work_type',
 'smoking_status']

target = ['stroke']

X = database[features]
y = database[target]
```

```python
X.isnull().sum()
```

```python
#preenchendo a linha que está com um dado nulo
X['gender'].fillna(1)
```

```python
#separando o database em dados de treino e de teste
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.25, random_state = 5)
```

```python
#padronizando os dados de treino e teste
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import plot_roc_curve
```

```python
#utilizando Regressão Logistica como classificador
#criando o modelo

lrModel = LogisticRegression()
lrModel.fit(X_train, y_train)
y_pred = lrModel.predict(X_test)

score = accuracy_score(y_test, y_pred)
print("O 'accuracy score' foi de:",score)
cMatrix = confusion_matrix(y_test, y_pred, labels = [1,0])
print(cMatrix)
plot_confusion_matrix(lrModel, X_test, y_test, labels = [1,0], cmap = "YlGnBu")
```

```python
plot_roc_curve(lrModel, X_test, y_test)
```

```python
#utilizando Árvore de Decisão como classificador
#criando o modelo
dtModel = DecisionTreeClassifier()
dtModel.fit(X_train, y_train)
y_pred = dtModel.predict(X_test)

score = accuracy_score(y_test, y_pred)

print("O 'accuracy score' foi de: ",score)
cMatrix = confusion_matrix(y_test, y_pred, labels=[1,0])
print(cMatrix)
plot_confusion_matrix(dtModel, X_test, y_test, labels=[1,0], cmap = "YlGnBu")
```

```python
plot_roc_curve(dtModel, X_test, y_test)
```
